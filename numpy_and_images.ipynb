## Images

```python
# assign picture to variable
pic = Image.open('DATA/tt.jpg')
```

```python
# picture to numpy array
pic_arr = np.asarray(pic)
pic_arr.shape
```

```python
# imshow uses its own color schale based on the rgb values, so it might not be accurate to its representation
plt.imshow(pic_arr)
```

```python
# only red color from rgb, using gray colormap
pic_red = pic_arr.copy()
plt.imshow(pic_red[:,:,0], cmap='gray')
```

```python
# only show red by assigning zero to green and blue values
pic_red_only = pic_arr.copy()
# get rid of green
pic_red_only[:,:,1] = 0
# get rid of blue
pic_red_only[:,:,2] = 0
# show
plt.imshow(pic_red_only)
```
