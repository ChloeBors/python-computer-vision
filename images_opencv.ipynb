## Image Basics

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
# opencv reads directly as numpy array
# opencv uses different order for colors 
# matplotlib -> RGB
# opencv -> BGR
img = cv2.imread('DATA/tt.jpg')
type(img)
```

```python
# change color order
fixed_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(fixed_img)
```

```python
# read an image in grayscale
img_gray = cv2.imread('DATA/tt.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray, cmap='gray')
```

```python
# resize image (width / height)
plt.imshow(cv2.resize(fixed_img, (1000, 200)))
```

```python
# resize by ratio
width_ratio = 0.05
height_ratio = 0.05
pixel_tt = cv2.resize(fixed_img, (0,0), fixed_img, width_ratio, height_ratio)
plt.imshow(pixel_tt)
```

```python
# flip
plt.imshow(cv2.flip(fixed_img,1))
```

```python
# save
cv2.imwrite('pixel_tt.jpg', pixel_tt)
```

## Open images outside of Jupyter with OpenCV

```python
im = cv2.imread('DATA/tt.jpg')
cv2.imshow('TT', img)

# close image with escape key
k = cv2.waitKey(0)
# if any other key is pressed; change nothing
while k != 27:
    k = cv2.waitKey(0)
# else, if the escape key is pressed, close the window
else:
    cv2.destroyAllWindows()
```
